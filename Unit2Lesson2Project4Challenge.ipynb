```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('total_waterborne_commerce.csv')
```

```python
df
```

```python
plt.figure(figsize=(14,14))

plt.subplot(2,2,1)
plt.scatter(x=df['Year'], y=df['Total'], color='red')
plt.ylabel('Tonnage (Billions)')
plt.xlabel('Year (1974-2013)')
plt.title('Total Tonnage Carried')

plt.subplot(2,2,2)
plt.plot(df['Year'], df['Foreign'], color='blue')
plt.ylabel('Tonnage (Billions)')
plt.xlabel('Year (1974-2013)')
plt.title('Foreign Tonnage Carried')

plt.subplot(2,2,3)
plt.plot(df['Year'], df['Domestic'], color='green')
plt.ylabel('Tonnage (Billions)')
plt.xlabel('Year (1974-2013)')
plt.title('Domestic Tonnage Carried')

plt.subplot(2,2,4)
plt.plot(df['Year'], df['Total'], color='red')
plt.plot(df['Year'], df['Foreign'], color='blue')
plt.plot(df['Year'], df['Domestic'], color='green')
plt.xlabel('Year (1974-2013)')
plt.title('Tonnage Carried: Foreign, Domestic & Total')

plt.show()
```

This series of data visualizations displays the total tonnage of commodities, both foreign and domestic, carried on United States waterways between 1974 and 2013. The data was assembled by the United States Army Corps of Engineers, and is available at data.gov as a .csv (Comma Separated Value) file.

The first visualization is a scatter plot with total tonnage in each year represented as a red dot on the y-axis, and the year on the x-axis. The tonnage appears to rise from the mid- to late- 1970s, then come back down to 1974 levels by 1983, only to rise again by a factor of three or four by 2013.

The second visualization is a line plot with foreign tonnage of commodities carried, using a blue color for the line. Once again, the tons carried are on the y-axis and the years are on the x-axis.

The third visualization is a line plot with domestic tonnage of commodities carried, using a green color for the line. The y-axis has a smaller scale (in billions of tons carried) than the y-axis for foreign tonnage, because the total variation of domestic tonnage was smaller than for foreign tonnage in the years contained in the data set.

The fourth visualization is a line plot of all three tonnage columns in the data set - foreign, domestic, and total tons of commodities carried (in billions) on U.S. waterways in the years 1974-2013. Foreign tonnage began to rise in 1983 along with domestic tonnage, but rose more rapidly, and since 1994 has constituted a larger proportion of waterborne commerce in the United States than domestic tonnage. Domestic tonnage fell more dramatically during the 2008 economic downturn than foreign tonnage.
